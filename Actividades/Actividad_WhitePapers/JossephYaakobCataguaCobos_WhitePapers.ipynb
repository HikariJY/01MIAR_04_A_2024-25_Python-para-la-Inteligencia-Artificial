```python
# Librerias requeridas para los ejemplos
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
import math
## Para el artículo 01
import tensorflow as tf
## Para el artículo 02
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Generar datos sintéticos para los dos ejemplos
np.random.seed(0)  # Semilla para reproducibilidad
x_train = np.linspace(0, 10, 100).reshape(-1, 1)  # Crear una secuencia de 100 valores de 0 a 10 y redimensionar
y_train = 3 * x_train + np.random.normal(0, 1, x_train.shape)  # Generar valores de salida con un poco de ruido
```

## Resumen del artículo 01 "TensorFlow: A system for large-scale machine learning"
El artículo presenta **TensorFlow** como un sistema de código abierto para el aprendizaje automático a gran escala. Se destaca la capacidad de **TensorFlow** para construir y entrenar modelos de aprendizaje automático en una variedad de plataformas, desde dispositivos móviles hasta sistemas distribuidos de alta gama. **TensorFlow** utiliza un grafo computacional para representar los cálculos del modelo, lo que permite la optimización automática y la distribución de la ejecución en múltiples *CPUs* o *GPUs*. Además, **TensorFlow** ofrece una amplia gama de herramientas y bibliotecas que facilitan la construcción y el entrenamiento de modelos de aprendizaje automático. [1]

### Ejemplo de código en Python:
A continuación, se muestra unos ejemplos simplse de cómo utilizar **TensorFlow** para realizar una operación matemática básica:

#### Ejemplo: Modelo de regresión lineal
En este ejemplo, primero generamos datos sintéticos para entrenar nuestro modelo de regresión lineal. Luego, definimos el modelo de regresión lineal utilizando la *API* funcional de *Keras*. Compilamos el modelo especificando el optimizador (Descenso de gradiente estocástico) y la función de pérdida (error cuadrático medio). Entrenamos el modelo utilizando los datos de entrada y salida generados anteriormente durante $100$ épocas. Finalmente, obtenemos los pesos finales del modelo y graficamos los datos de entrenamiento junto con la línea de regresión ajustada. [1]

```python
# Definir el modelo de regresión lineal
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),  # Definir la capa de entrada con una sola característica
    tf.keras.layers.Dense(units=1)       # Definir una capa densa con un solo nodo (regresión lineal)
])

model.compile(optimizer='sgd', loss='mse')  # Compilar el modelo con descenso de gradiente estocástico (sgd) y error cuadrático medio como función de pérdida (mse)
history = model.fit(x_train, y_train, epochs=100, verbose=0)  # Entrenar el modelo con los datos de entrada y salida durante 100 épocas (verbose - 0 para nada - 1 para ver una barra animada de progreso - 2 para ver datos de entrenamiento)
w_final, b_final = model.layers[0].get_weights()  # Obtener los pesos (pendiente) y sesgo finales del modelo

w_final_1 = np.squeeze(w_final)
b_final_1 = np.squeeze(b_final)

# Graficar los datos y la línea de regresión
plt.figure(figsize=(6, 4))
plt.scatter(x_train, y_train, label='Datos de entrenamiento')  # Graficar los datos de entrenamiento
plt.plot(x_train, np.squeeze(w_final_1) * x_train + b_final_1, color='red', label='Linea de regresion ajustada')  # Graficar la línea de regresión ajustada
plt.legend()  # Mostrar leyenda
plt.show()  # Mostrar gráfico
```

## Resumen del artículo 02 "PyTorch: An Imperative Style, High-Performance Deep Learning Library"
El artículo presenta **PyTorch**, una biblioteca de aprendizaje profundo que proporciona una forma imperativa y de alto rendimiento para construir modelos de aprendizaje profundo. A diferencia de otras bibliotecas que utilizan una definición estática de gráficos computacionales (como **TensorFlow** antes de la versión 2.0), PyTorch permite definir gráficos dinámicos, lo que facilita la construcción y depuración de modelos. 3. ***Integración con Python:*** **PyTorch** se integra profundamente con Python, lo que facilita el uso de todas las bibliotecas y herramientas del ecosistema Python.

5. ***Alto Rendimiento:*** **PyTorch** está diseñado para ser eficiente y aprovechar al máximo el hardware disponible, incluyendo *GPU* y *TPU*.

9. ***Comunidad y Ecosistema:*** **PyTorch** tiene una comunidad activa y un ecosistema en crecimiento, con muchas herramientas y bibliotecas construidas sobre **PyTorch** para diversas aplicaciones en el aprendizaje profundo.

### Ejemplo de código en PythonA continuación, se muestra unos ejemplos simplse de cómo utilizar **PytTorch** para realizar una operación matemática básica:.

#### Ejemplo: Modelo de regresión lineal.
En este ejemplo, primero generamos datos sintéticos para entrenar nuestro modelo de regresión lineal. Luego, definimos el modelo de regresión lineal utilizando **PyTorch**. Definimos una clase de modelo que hereda de `nn.Module` y utilizamos una capa lineal. Compilamos el modelo especificando el optimizador (Descenso de gradiente estocástico) y la función de pérdida (error cuadrático medio). Entrenamos el modelo utilizando los datos de entrada y salida generados anteriormente durante $100$ épocas. Finalmente, obtenemos los pesos finales del modelo y graficamos los datos de entrenamiento junto con la línea de regresión ajustada. [2]

```python
# Convertir a tensores de PyTorch
x_train_tensor = torch.from_numpy(x_train).float()  # Convertir datos de entrada a tensor de PyTorch
y_train_tensor = torch.from_numpy(y_train).float()  # Convertir datos de salida a tensor de PyTorch

# Definir el modelo de regresión lineal
class LinearRegressionModel(nn.Module):
    def __init__(self): # Definición del modelo
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(1, 1)  # Capa lineal con una entrada y una salida

    def forward(self, x): # Propagación hacia adelante
        return self.linear(x)

model = LinearRegressionModel()  # Instanciar el modelo

# Definir la función de pérdida y el optimizador
criterion = nn.MSELoss()  # Error cuadrático medio como función de pérdida
optimizer = optim.SGD(model.parameters(), lr=0.01)  # Descenso de gradiente estocástico (SGD) con tasa de aprendizaje de 0.01

# Entrenar el modelo
num_epochs = 100  # Número de pasos
for epoch in range(num_epochs):
    model.train()  # Modo de entrenamiento
    outputs = model(x_train_tensor)  # Obtener las predicciones del modelo
    loss = criterion(outputs, y_train_tensor)  # Calcular la pérdida
    
    optimizer.zero_grad()  # Limpiar los gradientes del optimizador
    loss.backward()  # Calcular los gradientes
    optimizer.step()  # Actualizar los parámetros del modelo

    if (epoch+1) % 10 == 0:  # Imprimir la pérdida cada 10 épocas
        print(f'Pasos [{epoch+1}/{num_epochs}], Pérdida: {loss.item():.4f}')

w_final_2, b_final_2 = model.linear.weight.item(), model.linear.bias.item() # Obtener los pesos finales del modelo
# Obtener las predicciones del modelo
model.eval()  # Modo de evaluación
predicted = model(x_train_tensor).detach().numpy()  # Obtener las predicciones y convertir a numpy

# Graficar los datos y la línea de regresión
plt.figure(figsize=(6, 4))
plt.scatter(x_train, y_train, label='Datos de entrenamiento')  # Graficar los datos de entrenamiento
plt.plot(x_train, predicted, color='red', label='Línea de regresión ajustada')  # Graficar la línea de regresión ajustada
plt.legend()  # Mostrar leyenda
plt.show()  # Mostrar gráfico
```

## Resultados

### Resultados para TensorFlow

```python
# Imprimir los pesos finales
print(f'Pesos finales: {w_final_1}.')
print(f'Sesgo final: {b_final_1}.')
print(f'Ecuación de la recta: y={w_final_1:.4f}x+{b_final_1:.4f}.')
print()
# Prueba de datos
index_value = 50
print(f'Valor de entrada: {x_train[index_value,0]:.4f}.')
print(f'Valor de salida esperada: {y_train[index_value,0]:.4f}')
y_value = w_final_1 * x_train[index_value] + b_final_1
print(f'Valor de salida calculada: {y_value[0]:.4f}')
print()
# Error de salida
print(f'Error de salida:')
print(f'Error Absoluto Medio (MAE): {mean_absolute_error(y_train[index_value], y_value):.4f}') # Error Absoluto Medio (MAE)
print(f'Error Cuadrático Medio (MSE): {mean_squared_error(y_train[index_value], y_value):.4f}') # Error Cuadrático Medio (MSE)
mape = np.mean(np.abs((y_train[index_value] - y_value) / y_train[index_value])) * 100
print(f'Error Porcentual Medio Absoluto (MAPE): {mape:.4f}%')
```

### Resultados para PyTorch

```python
# Imprimir los pesos finales
print(f'Pesos finales: {w_final_2}.')
print(f'Sesgo final: {b_final_2}.')
print(f'Ecuación de la recta: y={w_final_2:.4f}x+{b_final_2:.4f}.')
print()
# Prueba de datos
# index_value = 50
print(f'Valor de entrada: {x_train[index_value,0]:.4f}.')
print(f'Valor de salida esperada: {y_train[index_value,0]:.4f}')
y_value = w_final_2 * x_train[index_value] + b_final_2
print(f'Valor de salida calculada: {y_value[0]:.4f}')
print()
# Error de salida
print(f'Error de salida:')
print(f'Error Absoluto Medio (MAE): {mean_absolute_error(y_train[index_value], y_value):.4f}') # Error Absoluto Medio (MAE)
print(f'Error Cuadrático Medio (MSE): {mean_squared_error(y_train[index_value], y_value):.4f}') # Error Cuadrático Medio (MSE)
mape = np.mean(np.abs((y_train[index_value] - y_value) / y_train[index_value])) * 100
print(f'Error Porcentual Medio Absoluto (MAPE): {mape:.4f}%')
```

### Gráfica de comparación entre TensorFlow y PyTorch
